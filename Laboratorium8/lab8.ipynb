```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint 
```

```python
#stale

N = 19000
beta = 0.65
gamma = 0.2

# warunki poczatkowe

S_0 = 19000
I_0 = 5
R_0 = 0
```

Model SIR 
$$S \rightarrow I \rightarrow R $$

Dynamika modelu
$$\frac{dS}{dt} = - \beta \frac{S(t)I(t)}{N} $$
$$ \frac{dI}{dt} = \beta \frac{S(t)I(t)}{N} - \gamma I(t) $$
$$ \frac{dR}{dt} = \gamma I(t) $$

Gdzie:
$$ \beta - \text{współczynnik zarażenia} $$ 
$$ \gamma - \text{współczynnik zdrowienia} $$

Definiuje wektor 
$$ \vec{y} = [S,I,R]$$
$$\frac{d\vec{y}}{dt} = [-\beta \frac{SI}{N}, \beta \frac{SI}{N} - \gamma I, \gamma I]

```python
#funkcja na pochodna

def funkcja_pochodna(y, t):
    S, I, R = y

    return -beta * S * I / N, beta * S * I / N - gamma * I, gamma * I
```

```python
#kroki czasowe 

t = np.linspace(0, 100, 1000)
```

```python
#rozwiazanie ukladu rownan

wyniki = odeint(funkcja_pochodna, (S_0, I_0, R_0), t)
```

```python
plt.plot(t, wyniki[:, 0], 'g', label = 'S')
plt.plot(t, wyniki[:, 1], 'r', label = 'I') 
plt.plot(t, wyniki[:, 2], 'b', label = 'R')

plt.ylabel("Ile osob w danym stanie")
plt.xlabel("Czas t")
plt.title("Model SIR")
plt.legend()
plt.show()
```
