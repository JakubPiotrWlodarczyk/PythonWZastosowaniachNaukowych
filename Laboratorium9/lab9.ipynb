```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint 
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.layouts import row, column, gridplot, layout
from bokeh.models import Slider, Div

output_notebook()
```

```python
#stale

N = 19000
beta = 0.65
gamma = 0.2

# warunki poczatkowe

S_0 = 19000
I_0 = 5
R_0 = 0
```

```python
#funkcja na pochodna

def funkcja_pochodna(y, t):
    S, I, R = y

    return -beta * S * I / N, beta * S * I / N - gamma * I, gamma * I
```

```python
#kroki czasowe 

t = np.linspace(0, 100, 1000)
```

```python
#rozwiazanie ukladu rownan

wyniki = odeint(funkcja_pochodna, (S_0, I_0, R_0), t)
```

```python
fig = figure(sizing_mode='stretch_width', 
            aspect_ratio=2,
            title='Model SIR',
            x_axis_label='Czas t',
            y_axis_label='Ilo osb w danym stanie')
fig.toolbar.logo = None 
fig.toolbar.autohide = True

fig.line(t, wyniki[:,0], color='green', line_width=2, legend_label='S')
fig.line(t, wyniki[:,1], color='red', line_width=2, legend_label='I')
fig.line(t, wyniki[:,2], color='blue', line_width=2, legend_label='R')

show(fig)
```

```python
slider_N = Slider(start = 0, end = 100000, step = 1000, value = 19000, title = 'N', sizing_mode='stretch_width')
slider_N.js_link('value', )

show(column(Div(text='Liczba osobnikw'), row(column(slider_N, width=200), fig, width=1000)))
```
